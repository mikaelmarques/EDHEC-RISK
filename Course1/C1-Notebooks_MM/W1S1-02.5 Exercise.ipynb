```python
#Let's import the data and try to assess sharpe ratio for the 2nd and 4th quantile.
import pandas as pd
returns = pd.read_csv('W1S1 data/Portfolios_Formed_on_ME_monthly_EW.csv',
                     na_values = -99.99, index_col = 0, parse_dates = True)
returns.head()
```

```python
returns = returns[['Qnt 2', 'Qnt 4']]
returns.columns = ['2nd Quantile', '4th Quantile']
returns = returns/100
returns.head()
```

```python
#We have monthly data that we have to annualize to find the annualized total return
ann_return = (returns + 1).prod()**(12/returns.shape[0]) - 1
ann_vol = returns.std()*12**0.5
rf = 0.03
sharpe_ratio = (ann_return - rf)/ann_vol
sharpe_ratio
```
