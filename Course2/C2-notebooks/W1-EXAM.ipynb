```python
import pandas as pd
import numpy as np
import edhec_risk_kit_204 as erk
```

```python
data = erk.get_ind_file('returns', n_inds=49)
data = data['1991':]
data.head()
```

```python
factor = erk.get_fff_returns()
factor = factor['1991':]
factor.head()
```

```python
excess_ret_beer = data['Beer'] - factor['RF'].values
excess_ret_beer = pd.DataFrame(excess_ret_beer)
exp_var = factor['Mkt-RF'].copy()
exp_var = pd.DataFrame(exp_var)
lm = erk.regress(excess_ret_beer, exp_var, alpha=True)
lm.summary()
```

```python
excess_ret_steel = data['Steel'] - factor['RF'].values
excess_ret_steel = pd.DataFrame(excess_ret_steel)
lm = erk.regress(excess_ret_steel, exp_var, alpha=True)
lm.summary()
```

```python
excess_ret_beer_2013 = data.loc['2013':,['Beer']] - factor.loc['2013':,['RF']].values
excess_ret_beer_2013 = pd.DataFrame(excess_ret_beer_2013)
exp_var = factor.loc['2013':,['Mkt-RF']].copy()
exp_var = pd.DataFrame(exp_var)
lm = erk.regress(excess_ret_beer_2013, exp_var, alpha=True)
lm.summary()
```

```python
excess_ret_steel_2013 = data.loc['2013':,['Steel']] - factor.loc['2013':,['RF']].values
excess_ret_steel_2013 = pd.DataFrame(excess_ret_steel_2013)
lm = erk.regress(excess_ret_steel_2013, exp_var, alpha=True)
lm.summary()
```
