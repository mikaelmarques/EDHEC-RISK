```python
import pandas as pd
import numpy as np
import edhec_risk_kit_206 as erk

%load_ext autoreload
%autoreload 2

ind_rets = erk.get_ind_returns(n_inds=49)
ind_rets = ind_rets['2014':'2018']

ind_marketcap = erk.get_ind_market_caps(n_inds=49)
ind_marketcap = ind_marketcap['2014':'2018']
```

```python
ind_cov = ind_rets.cov()

capweighted = erk.weight_cw(ind_rets, ind_marketcap)
rc_capweighted = erk.risk_contribution(capweighted, ind_cov)
rc_capweighted = rc_capweighted.sort_values(ascending=False)
rc_capweighted
```

```python
equalweighted = erk.weight_ew(ind_rets)
rc_ew = erk.risk_contribution(equalweighted, ind_cov)
rc_ew = rc_ew.sort_values(ascending=False)
rc_ew
```
